# **GENERATING IMAGES USING STABLE DIFUSION PRE-TRAINED MODELS**

```python
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
import gradio as gr

# Load the pretrained Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
pipe = pipe.to("cuda")  # Ensure it's on the GPU for fast inference
```

```python

def generate_image(prompt):
    with autocast("cuda"):
        image = pipe(prompt).images[0]
    return image

# Gradio interface for the image generation app
gr_interface = gr.Interface(
    fn=generate_image,
    inputs=gr.Textbox(lines=2, placeholder="Enter your image prompt here..."),  # Updated API
    outputs=gr.Image(type="pil"),  # Updated API
    title="Stable Diffusion Image Generator",
    description="Enter a prompt to generate an image using Stable Diffusion."
)

# Launch the app
gr_interface.launch(debug=True)
```
